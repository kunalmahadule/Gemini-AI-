# Multimodel Project (Image to text Generation)

```python
import PIL.Image
import PIL.ImageChops
import google.generativeai as genai

img = PIL.Image.open(r"C:\Users\GauravKunal\Desktop\multi-ethnic-group-of-men-standi.jpg")
img
```

```python
# model = genai.GenerativeModel('models/gemini-2.0-pro-exp-02-05')

model = genai.GenerativeModel('models/gemini-2.0-flash-exp-image-generation')
```

```python
response = model.generate_content(img) # Giving image to the model

# to_markdown(response.text)
# print(response.text) --> this will show response in one line
text = response.text
wrapped_text = textwrap.wrap(text, width=100)
print('\n'.join(wrapped_text))
```

```python
# Ensure the image is loaded correctly before passing it to the model
try:
    img = PIL.Image.open(r"C:\Users\GauravKunal\Desktop\multi-ethnic-group-of-men-standi.jpg")
    response = model.generate_content(img)  # Pass the image to the model

    text = response.text
    wrapped_text = textwrap.wrap(text, width=100)
    print('\n'.join(wrapped_text))
except FileNotFoundError:
    print("The image file is missing. Please provide a valid image file.")
except PIL.UnidentifiedImageError:
    print("The image file is invalid or corrupted.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Ask the model a question based on the image
try:
    img = PIL.Image.open(r"C:\Users\GauravKunal\Desktop\multi-ethnic-group-of-men-standi.jpg")
    # response = model.generate_content("How many number of people are in the image?", image=img)  # Pass the image and question to the model

    response = model.generate_image_content(prompt="How many number of people are in the image?", image=img)
    
    text = response.text
    wrapped_text = textwrap.wrap(text, width=100)
    print('\n'.join(wrapped_text))
except FileNotFoundError:
    print("The image file is missing. Please provide a valid image file.")
except PIL.UnidentifiedImageError:
    print("The image file is invalid or corrupted.")
except Exception as e:
    print(f"An error occurred: {e}")
```
